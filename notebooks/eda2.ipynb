Pdfplumber + Tesseract OCR / hwp5txt 실험

```python
import json
from dataclasses import dataclass, field
from typing import Optional

@dataclass
class DocumentChunk:
    chunk_id: str
    공고번호: str
    사업명: str
    발주기관: str
    사업금액: Optional[float] = None
    입찰마감일: Optional[str] = None
    파일형식: str = "hwp"
    text_source: str = "file_extract"
    text: str = ""
    metadata: dict = field(default_factory=dict)

chunks_list = []
jsonl_path = "../data/rfp_chunks.jsonl"  # 노트북 위치에 따라 경로 확인!

with open(jsonl_path, "r", encoding="utf-8") as f:
    for line in f:
        if line.strip():
            chunks_list.append(DocumentChunk(**json.loads(line)))

print(f"🚀 OCR 데이터가 모두 포함된 {len(chunks_list):,}개의 완벽한 청크셋 로드 완료!")
```

```python
import pandas as pd
from dataclasses import asdict

# dataclass 리스트를 Pandas DataFrame으로 깔끔하게 변환
df_chunks = pd.DataFrame([asdict(c) for c in chunks_list])

print("─── [1] 전체 청크셋 기본 정보 ───")
print(df_chunks.info())

print("\n─── [2] PDF vs HWP 청크 비율 ───")
print(df_chunks['파일형식'].value_counts(normalize=True) * 100)

print("\n─── [3] 가장 많은 청크가 생성된 상위 5개 사업 ───")
print(df_chunks['사업명'].value_counts().head())
```

```python
import random

# 전체 청크 중 랜덤하게 3개 뽑아서 출력
sampled_chunks = random.sample(chunks_list, 3)

for i, chunk in enumerate(sampled_chunks, 1):
    print(f"=== [샘플 {i}] {chunk.사업명} ({chunk.파일형식.upper()}) ===")
    print(f"• 청크 ID   : {chunk.chunk_id}")
    print(f"• 발주기관  : {chunk.발주기관}")
    print(f"• 추출 출처 : {chunk.text_source}")
    print(f"• 본문 요약 (앞 300자) :\n{chunk.text[:300]}...")
    print("-" * 60)
```

```python
import pandas as pd
from dataclasses import asdict

# 1. 원본 데이터프레임 로드
df_all = pd.DataFrame([asdict(c) for c in chunks_list])

# 2. [도메인 필터링 조건 설정]
# 문제가 된 '아세안+3...' 사업이면서 'hwp' 파일에서 파싱된 청크들을 노이즈 타깃으로 지정합니다.
is_target_garbage = (df_all['사업명'].str.contains('아세안')) & (df_all['파일형식'] == 'hwp')

# 3. 타깃 노이즈를 제외한 깨끗한 데이터와 쓰레기 데이터 분리
df_clean = df_all[~is_target_garbage].copy()
df_garbage = df_all[is_target_garbage].copy()

print("─── 🎯 도메인 타격 기반 필터링 결과 ───")
print(f"📦 필터링 전 전체 청크 : {len(df_all):,}개")
print(f"✨ 필터링 후 가용한 청크: {len(df_clean):,}개 (약 1,981개 예상)")
print(f"🗑️ 제거된 외계어 노이즈  : {len(df_garbage):,}개 (약 1,411개 예상)")

print("\n─── ✨ 외계어 소탕 후 진짜 상위 5개 사업 분포 ───")
print(df_clean['사업명'].value_counts().head())

# 4. 최종 리스트 복원
final_chunks_list = [
    DocumentChunk(
        chunk_id=r['chunk_id'], 공고번호=r['공고번호'], 사업명=r['사업명'], 발주기관=r['발주기관'],
        사업금액=r['사업금액'], 입찰마감일=r['입찰마감일'], 파일형식=r['파일형식'],
        text_source=r['text_source'], text=r['text'], metadata=r['metadata']
    )
    for _, r in df_clean.iterrows()
]
```

```python
import os
from pathlib import Path

# 1. 실제 리눅스 서버 데이터 폴더에 있는 파일 목록 (확장자가 hwp, pdf인 것만)
# 폴더 경로는 예원님의 실제 files 폴더 경로에 맞게 지정하세요.
DATA_DIR = Path("../data/files") 
actual_files = set([f.name for f in DATA_DIR.glob("*") if f.suffix.lower() in ['.hwp', '.pdf']])

# 2. 파싱 완료된 df_clean 데이터에 존재하는 파일 목록
parsed_files = set(df_clean['파일명'].unique()) if '파일명' in df_clean.columns else set()

# 만약 df_clean에 파일명 컬럼이 없다면 metadata 딕셔너리나 파일 경로 문자열에서 추출해야 합니다.
# (앞서 제공된 데이터프레임 구조에 '파일명' 정보가 있다면 아래 코드가 정확히 작동합니다)
if not parsed_files and 'text_source' in df_clean.columns:
    # chunk_id나 다른 메타데이터를 활용해 파싱된 파일명을 유추하는 백업 로직
    # 여기서는 예원님의 df_clean 구조에 맞춰 매칭합니다.
    pass

print("─── 📊 파일 파싱 누락 여부 최종 점검 ───")
print(f"📁 실제 폴더 내 원본 파일 개수 : {len(actual_files)}개")
print(f"✨ 정상 파싱 완료된 파일 개수   : {len(parsed_files)}개")

# 3. 누락된 파일 찾아내기
missing_files = actual_files - parsed_files

if len(missing_files) == 0:
    print("\n✅ 대성공! 원본 파일 100%가 누락 없이 전부 파싱되어 청크셋에 반영되었습니다.")
else:
    print(f"\n⚠️ 경고: {len(missing_files)}개의 파일이 파싱에서 누락되었습니다!")
    print("❌ 누락된 파일 명단:")
    for f in missing_files:
        print(f"  - {f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 깨짐 방지 설정 (리눅스 서버 환경에 맞는 폰트 설정이 필요할 수 있습니다)
plt.rc('font', family='NanumGothic' if os.path.exists('/usr/share/fonts/truetype/nanum/NanumGothic.ttf') else 'sans-serif')

# 파일별 청크 개수 집계
chunk_counts_per_file = df_clean['사업명'].value_counts()

print("─── 📈 파일(사업)별 생성된 청크 개수 요약 ───")
print(chunk_counts_per_file)

# 청크 개수가 0이거나 극단적으로 적은(예: 2개 이하) 문서가 있는지 확인
underparsed_files = chunk_counts_per_file[chunk_counts_per_file <= 2]
if len(underparsed_files) > 0:
    print("\n💡 [확인 필요] 아래 문서들은 규모에 비해 청크가 너무 적게 생성되었습니다. 본문 공백 상태를 확인해 보세요:")
    print(underparsed_files)
```

```python
# chunk_id 내에 페이지를 뜻하는 '_p' 패턴이 확실하게 박힌 청크 정밀 필터링
korea_univ_chunks = [
    c for c in final_chunks_list 
    if c.발주기관 == "고려대학교" and "_p" in c.chunk_id
]

print(f"📂 고려대학교 이미지/OCR 파싱 청크 개수: {len(korea_univ_chunks)}개")
print("-" * 60)

if len(korea_univ_chunks) > 0:
    # 대표 샘플 상위 2개 추출하여 원본 이미지 속 글자가 텍스트로 복원됐는지 확인
    for i, chunk in enumerate(korea_univ_chunks[:2], 1):
        print(f"[{i}] 청크 ID: {chunk.chunk_id}")
        print(f" 출처 엔진: {chunk.text_source} | 파일 포맷: {chunk.파일형식.upper()}")
        print(f"📄 추출 본문 스냅샷 (앞 400자):\n{chunk.text[:400]}")
        print("-" * 60)
else:
    print("💡 고려대학교 OCR 지정 청크를 찾을 수 없습니다. 메타데이터 조건을 확인해 주세요.")
```
