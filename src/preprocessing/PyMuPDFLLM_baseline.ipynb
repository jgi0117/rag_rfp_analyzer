### PyMuPDF4LLM

```python
import pymupdf4llm
import time

# filepath: 파일 경로
# pages → 9페이지만: [8], 전체: None
# image_path: 이미지 저장 경로
# write_images → True: 이미지 추출, False: 이미지 추출X

def extract_pdf(filepath, pages=None, image_path="./images", write_images=False):
    md_text = pymupdf4llm.to_markdown(
        filepath,
        pages=pages,
        write_images=write_images,
        image_path=image_path,
    )
    return md_text
```

```python
# 8페이지만 확인 
filepath = "../../data/files/고려대학교_차세대 포털·학사 정보시스템 구축사업.pdf"

# 시간 측정
start = time.time()
md_text = extract_pdf(filepath, pages=[8])
elapsed = time.time() - start

print(md_text)
print(f"\n처리 시간: {elapsed:.3f}초")
```
